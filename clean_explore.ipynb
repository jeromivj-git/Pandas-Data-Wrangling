```python
import pandas as pd
import numpy as np

# =====================================================
# LOAD DATASET FROM GITHUB
# =====================================================

url = "https://github.com/jeromivj-git/Customer_Details/raw/refs/heads/main/customer_sales_dataset.xlsx"

df = pd.read_excel(url)

print("\nDATASET LOADED SUCCESSFULLY\n")

# =====================================================
# MENU DRIVEN PROGRAM
# =====================================================

while True:

    print("\n========================= MENU =========================")

    print("1. Display Full Dataset\t\t11. Delete Row")
    print("2. Head Function\t\t12. Set Custom Index")
    print("3. Tail Function\t\t13. Sort Dataset")
    print("4. Dataset Info\t\t\t14. Detect Missing Values")
    print("5. Describe Dataset\t\t15. Drop Missing Values")
    print("6. Value Counts\t\t\t16. Fill Missing Values")
    print("7. Explore Columns\t\t17. Group By and Aggregate")
    print("8. Add New Column\t\t18. Merge DataFrames")
    print("9. Update Column Values\t\t19. Concat DataFrames")
    print("10. Delete Column\t\t20. Join DataFrames")

    print("\n21. Save Cleaned Dataset")
    print("22. Exit")

    print("========================================================")

    choice = int(input("\nEnter Your Choice: "))

    # =====================================================
    # DISPLAY DATASET
    # =====================================================

    if choice == 1:

        print("\nFULL DATASET\n")
        print(df.to_string(index=False))

    # =====================================================
    # HEAD
    # =====================================================

    elif choice == 2:

        print("\nHEAD FUNCTION\n")
        print(df.head())

    # =====================================================
    # TAIL
    # =====================================================

    elif choice == 3:

        print("\nTAIL FUNCTION\n")
        print(df.tail())

    # =====================================================
    # INFO
    # =====================================================

    elif choice == 4:

        print("\nINFO FUNCTION\n")
        print(df.info())

    # =====================================================
    # DESCRIBE
    # =====================================================

    elif choice == 5:

        print("\nDESCRIBE FUNCTION\n")
        print(df.describe())

    # =====================================================
    # VALUE COUNTS
    # =====================================================

    elif choice == 6:

        print("\nVALUE COUNTS - CITY\n")
        print(df['City'].value_counts())

    # =====================================================
    # EXPLORE COLUMNS
    # =====================================================

    elif choice == 7:

        print("\nCOLUMN NAMES\n")
        print(df.columns)

        print("\nCUSTOMER NAMES\n")
        print(df['Customer_Name'])

        print("\nPRODUCT AND SALES\n")
        print(df[['Product', 'Sales']])

    # =====================================================
    # ADD COLUMN
    # =====================================================

    elif choice == 8:

        df['Bonus'] = df['Sales'] * 0.10

        print("\nBONUS COLUMN ADDED\n")
        print(df.head())

    # =====================================================
    # UPDATE VALUES
    # =====================================================

    elif choice == 9:

        df.loc[df['City'] == 'Chennai', 'Rating'] = 5

        print("\nUPDATED RATINGS FOR CHENNAI\n")
        print(df.head())

    # =====================================================
    # DELETE COLUMN
    # =====================================================

    elif choice == 10:

        col = input("Enter Column Name to Delete: ")

        if col in df.columns:

            df.drop(col, axis=1, inplace=True)

            print(f"\n{col} COLUMN DELETED\n")

        else:

            print("\nCOLUMN NOT FOUND")

    # =====================================================
    # DELETE ROW
    # =====================================================

    elif choice == 11:

        row = int(input("Enter Row Index to Delete: "))

        if row in df.index:

            df.drop(row, axis=0, inplace=True)

            print(f"\nROW {row} DELETED\n")

        else:

            print("\nROW NOT FOUND")

    # =====================================================
    # SET INDEX
    # =====================================================

    elif choice == 12:

        if 'Customer_ID' in df.columns:

            df.set_index('Customer_ID', inplace=True)

            print("\nCUSTOM INDEX SET\n")
            print(df.head())

        else:

            print("\nCustomer_ID ALREADY SET AS INDEX")

    # =====================================================
    # SORT DATA
    # =====================================================

    elif choice == 13:

        sorted_df = df.sort_values(by='Sales', ascending=False)

        print("\nSORTED BY SALES\n")
        print(sorted_df.head())

    # =====================================================
    # MISSING VALUES
    # =====================================================

    elif choice == 14:

        print("\nMISSING VALUES\n")
        print(df.isnull().sum())

    # =====================================================
    # DROP NULL
    # =====================================================

    elif choice == 15:

        drop_df = df.dropna()

        print("\nAFTER DROPPING NULL VALUES\n")
        print(drop_df.head())

    # =====================================================
    # FILL NULL
    # =====================================================

    elif choice == 16:

        df['Sales'] = df['Sales'].fillna(df['Sales'].mean())

        print("\nAFTER FILLING NULL VALUES\n")
        print(df.head())

    # =====================================================
    # GROUP BY
    # =====================================================

    elif choice == 17:

        grouped = df.groupby('City')['Sales'].mean()

        print("\nAVERAGE SALES BY CITY\n")
        print(grouped)

    # =====================================================
    # MERGE
    # =====================================================

    elif choice == 18:

        region_data = pd.DataFrame({

            'City': ['Chennai', 'Bangalore', 'Hyderabad', 'Mumbai', 'Delhi'],

            'Region': ['South', 'South', 'South', 'West', 'North']

        })

        merged_df = pd.merge(df.reset_index(), region_data, on='City')

        print("\nMERGED DATAFRAME\n")
        print(merged_df.head())

    # =====================================================
    # CONCAT
    # =====================================================

    elif choice == 19:

        extra_df = pd.DataFrame({

            'Customer_ID': [151, 152],

            'Customer_Name': ['Arun', 'Bala'],

            'City': ['Chennai', 'Delhi'],

            'Product': ['Mobile', 'Laptop'],

            'Sales': [26000, 62000],

            'Rating': [4, 5]

        })

        concat_df = pd.concat([df.reset_index(), extra_df])

        print("\nCONCAT DATAFRAME\n")
        print(concat_df.tail())

    # =====================================================
    # JOIN
    # =====================================================

    elif choice == 20:

        region_data = pd.DataFrame({

            'City': ['Chennai', 'Bangalore', 'Hyderabad', 'Mumbai', 'Delhi'],

            'Region': ['South', 'South', 'South', 'West', 'North']

        })

        join_df = df.join(region_data.set_index('City'), on='City')

        print("\nJOIN DATAFRAME\n")
        print(join_df.head())

    # =====================================================
    # SAVE DATASET
    # =====================================================

    elif choice == 21:

        df.to_excel("cleaned_customer_sales.xlsx", index=False)

        print("\nCLEANED DATASET SAVED SUCCESSFULLY")

    # =====================================================
    # EXIT
    # =====================================================

    elif choice == 22:

        print("\nPROGRAM TERMINATED")
        break

    # =====================================================
    # INVALID CHOICE
    # =====================================================

    else:

        print("\nINVALID CHOICE")
```
